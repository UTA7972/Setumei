```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.python.keras import layers
from tensorflow.python.keras.callbacks import EarlyStopping
from vit_keras import vit, utils
import numpy
from google.colab import drive
import os

drive.mount('/content/drive')
# 現在のディレクトリを取得する
current_dir = os.getcwd()

# ディレクトリ内のファイルとフォルダーを表示する
for entry in os.scandir(current_dir):
    if entry.is_file():
        print('File:', entry.name)
    elif entry.is_dir():
        print('Directory:', entry.name)
```

```python
# データセットを読み込む
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()

# データセットを正規化する
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

print(numpy.shape(x_train))
print(numpy.shape(x_test))
print(numpy.shape(y_train))
print(numpy.shape(y_test))
```

```python
# モデルのパラメータを設定する
input_shape = (32, 32, 3)
num_classes = 10
batch_size = 64
num_epochs = 10
checkpoint_path = '/content/drive/MyDrive/pro/model'

# ViTモデルを構築する
vit_model = vit.vit_b16(
    image_size=32,
    activation="softmax",
    pretrained=True,
    include_top=True,
    classes=num_classes,
)

# モデルをコンパイルする
vit_model.compile(
    loss="sparse_categorical_crossentropy",
    optimizer="adam",
    metrics=["accuracy"],
)

# 早期終了のコールバックを設定する
early_stopping = EarlyStopping(monitor="val_loss", patience=5, mode="min")

# モデルをトレーニングする
history = vit_model.fit(
    x_train,
    y_train,
    batch_size=batch_size,
    epochs=num_epochs,
    validation_split=0.1,
    callbacks=[early_stopping],
)

# モデルを評価する
test_loss, test_acc = vit_model.evaluate(x_test, y_test)
print("Test loss:", test_loss)
print("Test accuracy:", test_acc)
```
